** Take top 50 words from Shakespeare (all 3 books) 

```python
filenames = [
    "shakespeare-caesar.txt",
    "shakespeare-hamlet.txt",
    "shakespeare-macbeth.txt"
]          # Taking all three shakespear file and putting it together
with open("result.txt", "w") as f:
    for filename in filenames:
        f.write(nltk.corpus.gutenberg.raw(filename))
        
```

```python
print(nltk.corpus.gutenberg.raw(filename))
```

```python
from collections import Counter
def content_text(text):
    stopwords = set(nltk.corpus.stopwords.words('english'))
    with_stp = Counter()
    for word in text:
        # update count off all words in the line that are in stopwords
        word = word.lower()
        if word in stopwords:
             with_stp.update([word])
    # return a list with top ten most common words from each
    return [k for k,_ in with_stp.most_common(50)]
print ("The top 50 words used in the shakespear texts are as follows : ")
print(content_text(nltk.corpus.gutenberg.words(filename)))
```
