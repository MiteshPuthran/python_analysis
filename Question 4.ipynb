** Total posts done by user121

```python
import re   # importing regular expression
from os.path import basename  #returns the base file from the given path
from os.path import splitext  # to exract extension from the file name
```

```python
totalcount=0
for user121 in nltk.corpus.nps_chat.fileids():
    postcount=0
    base=basename(user121)
    base=splitext(base)[0]   # seperating out the files with xml extensions
    file = nltk.corpus.nps_chat.raw(user121)
    for line in file.split('\n'):
        if line.strip().startswith('<Post '):   #finding the tag where user data is found
            if 'User121' in line:
                postcount+=1
    totalcount+=postcount
print("Total posts done by User121 is ",totalcount)    
```
