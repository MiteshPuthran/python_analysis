** 2. Find out how many words and how many unique words in file 'austen-persuasion.txt'.

```python
persuasion = len(nltk.corpus.gutenberg.words('austen-persuasion.txt')) #code to find out total words in the file austen-persuasion.txt
print ("The total number of words in austen-persuasion.txt is " + str(persuasion)) # we need to convert integer to string to display
```

```python
unique = len(set(nltk.corpus.gutenberg.words('austen-persuasion.txt'))) # code to find out the unique words in the file austen-persuasion.txt
print ('The total number of unique words in austen-persuasion.txt is '+ str(unique)) # converting the integer to string
```

**Calculate the frequency for each words and show the top 10 most frequent words and their frequency.

```python
file = nltk.corpus.gutenberg.words('austen-persuasion.txt')
import string
file = [word.strip(string.punctuation) for word in file] # code to remove the punctuation marks from the text file
file = [word.lower() for word in file] # converting all the text to lowercase to find exact count of a particuar word.

wordcount={}

for word in file:
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1
        
copy = []  # copying the contents into a new list
for k,v in wordcount.items():
    copy.append((v, k)) # appending the key and its value together


copy = sorted(copy, reverse=True) # sorting the values in the descending order
print ("The frequency of each word in the text file in descending order is as follows :")
for k in copy:
        print (k[1], k[0])

```
