## Question 7

** Use Name corpus, calculate the probability of a name being of a male or female based on the last character(s) of name.

```python
import nltk
```

```python
from nltk.corpus import names
names.fileids()
```

```python
print(nltk.corpus.names.raw('male.txt'))
```

```python
print(nltk.corpus.names.raw('female.txt'))
```

```python
file1=nltk.corpus.names.words('male.txt')
file2=nltk.corpus.names.words('female.txt')
```

```python
char = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
```

```python
for c in char:
    male=[]
    female=[]
    for m,f in zip(file1,file2):   # zip is used to concatenate multiple files having same structure
        if m.endswith(c):
            male.append(m)
        if f.endswith(c):
            female.append(f)
    total=len(male)+len(female)
    try:
        maleprobability=round(len(male)/total*100,0)
        femaleprobability=round(len(female)/total*100,0)
    except ZeroDivisionError:
        maleprobability=0
        femaleprobability=0
    
    print('[',c,'] male : ',maleprobability,'% female : ',femaleprobability,'%')
```
